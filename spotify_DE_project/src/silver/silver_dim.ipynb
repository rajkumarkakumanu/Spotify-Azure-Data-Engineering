```python
from pyspark.sql.functions import *
from pyspark.sql.types import *
```

# DimUser

```python
df_user = spark.readStream.format("cloudFiles")\
                          .option("cloudFiles.format","parquet")\
                          .option("cloudFiles.schemaLocation","abfss://silver@rajspotifystorage.dfs.core.windows.net/DimUser/checkpoint")\
                          .load("abfss://bronze@rajspotifystorage.dfs.core.windows.net/DimUser")
```

```python
df_user = df_user.withColumn("user_name", lower(col("user_name")))
```

```python
display(df_user)
```

```python
import os
import sys
new_path = os.path.join(os.getcwd(),'..','..')
sys.path.append(new_path)
from utils.transformations import reuse
```

```python
del_col = reuse()
df_user = del_col.dropcol(df_user,['_rescued_data'])
display(df_user)
```

```python
df_user = df_user.dropDuplicates(['user_id'])
display(df_user)
```

```python
df_user.writeStream.format('delta')\
                   .outputMode('append')\
                   .option("checkpointLocation",'abfss://silver@rajspotifystorage.dfs.core.windows.net/DimUser/checkpoint')\
                   .trigger(once=True)\
                   .option("path","abfss://silver@rajspotifystorage.dfs.core.windows.net/DimUser/data")\
                   .toTable("spotify_cata.silver.DimUser")
```

# DimTrack

```python
df_track = spark.readStream.format("cloudFiles")\
                        .option("cloudFiles.format","parquet")\
                        .option("cloudFiles.schemaLocation","abfss://silver@rajspotifystorage.dfs.core.windows.net/DimTrack/checkpoint")\
                        .option("schemaEvolutionMode", "addNewColumn")\
                        .load("abfss://bronze@rajspotifystorage.dfs.core.windows.net/DimTrack")
```

```python
df_track = df_track.withColumn("Duration_Flag", when(col('duration_sec')<150,"Low")\
                                               .when(col('duration_sec')<300,'Medium')\
                                                 .otherwise('High'))
display(df_track)
```

```python
df_track = df_track.withColumn("track_name",regexp_replace(col("track_name"),'-',' '))
```

```python
df_track = del_col.dropcol(df_track,['_rescued_data'])
```

```python
display(df_track)
```

```python
df_track.writeStream.format('delta')\
                    .outputMode('append')\
                    .option("checkpointLocation","abfss://silver@rajspotifystorage.dfs.core.windows.net/DimTrack/checkpoint")\
                    .trigger(once=True)\
                    .option('path','abfss://silver@rajspotifystorage.dfs.core.windows.net/DimTrack/data')\
                    .toTable("spotify_cata.silver.DimTrack")
```
